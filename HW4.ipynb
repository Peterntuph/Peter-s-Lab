```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

hos = pd.read_csv("https://quality.data.gov.tw/dq_download_csv.php?nid=15393&md5_url=30e1cace9cb832bfd42525d41f5b3ce8")
hos.head(5)
```

```python
#統計各醫事人員的數量(純粹好奇)
doc = hos['醫師'].values
den = hos['牙醫師'].values
chimed = hos['中醫師'].values
pharm = hos['藥師'].values
ot = hos['職能治療師'].values
pt = hos['物理治療師'].values
nurse = hos['護理師'].values
list1 = [sum(doc),sum(den),sum(chimed),sum(pharm),sum(ot),sum(pt),sum(nurse)]
print(list1)
```

```python
hos2 =  hos[(hos['型態別'] == '醫院')] 
```

```python
hos3 = hos2.loc[:, '機構名稱':'牙體技術生'].copy()
hos3['醫護比'] = (hos3.護理師 + hos3.護士) / (hos3.醫師 + hos3.中醫師 + hos3.牙醫師) 
```

```python
hos3.head()
```

```python
hos4 = hos3.sort_values(by = ['醫護比'], ascending=False)
```

```python
hos4[:20]
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
hos4[:20].plot.bar(x='機構名稱', y=['醫護比'])
plt.title('台灣醫護比前20高的醫院')
```
