```python
x = np.linspace(-6, 6, 100)
y = (x - 5) * (x) * (x - 2)*(x + 4) ### x**4 - 3*x**3ã€€- 26*x**2 + 40

plt.plot(x, y);
plt.xlabel("x");
plt.ylabel("y");
plt.title("plot of f(x)");
```

```python
def func(x): return (x - 5) * (x) * (x - 2)*(x + 4)
```

```python
def GD(x_start, epochs, lr):    
    site = []    
    x = x_start   
    site.append(x)
    for i in range(epochs):
        dx = 4*x**3 - 9*x**2 - 52*x
        x = x - lr*dx  
        site.append(x)
    return site[-1]
```

```python
t = GD(-4,50,0.01)
print(t)
```

```python
plt.plot(x, func(x), c='r')   
plt.scatter(t, func(t), c='b' )
```
